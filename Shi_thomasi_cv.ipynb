```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread(r"C:\Users\visha\Downloads\flat_chessboard.png")
gray = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(flat_chess,cmap="gray")
```

```python
# Apply Shi-Tomasi Corner Detection
corners = cv2.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01,minDistance=10) 
corners = np.int0(corners)
# Draw circles around the corners
for corner in corners:
    x,y = corner.ravel()
    cv2.circle(flat_chess, (x, y), 5, (255, 0, 255), -1)
    plt.imshow(flat_chess)
```
