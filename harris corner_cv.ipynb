```python
import cv2
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread(r"C:\Users\visha\Downloads\flat_chessboard.png")
gray = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(flat_chess,cmap="gray")
```

```python
dst = cv2.cornerHarris(gray, 2, 3, 0.04) # image, harris corner matrix size , kernalsize , k value 
```

```python
plt.imshow(dst,cmap="gray")
```

```python
dst = cv2.dilate(dst, None) # none is structure element 
plt.imshow(dst,cmap="gray")
```

```python
#add color slicing 
flat_chess[dst > 0.01 * dst.max()] = [0, 0, 255]
plt.imshow(flat_chess)
```
