```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r"C:\Users\visha\OneDrive\Desktop\lab1.jpg")
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
img2 = cv2.imread(r"C:\Users\visha\OneDrive\Desktop\lab2.jpeg")
gray2 = cv2.cvtColor(img2 ,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray2,cmap="gray")
```

```python
sift = cv2.SIFT_create()
#orb = cv2.ORB_create()
```

```python
# 
kp1,des1 = sift.detectAndCompute(gray1,None)
kp2,des2 = sift.detectAndCompute(gray2,None)
```

```python
# brute force- bf take all the matches 
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1,des2,k=2)
```

```python
good_matches =[]
for m,n in matches:
    if m.distance < 0.78 * n.distance:
        good_matches.append(m)
```

```python
good_matches
```

```python
match_img = cv2.drawMatches(gray1,kp1,gray2,kp2,good_matches,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(match_img)
```
