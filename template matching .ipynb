```python
import cv2
import matplotlib.pyplot as plt 
#template imag --> 
# all the template matching methods are correlation based methods 
#target --> emplate img
#actual
```

```python
img1 = cv2.imread(r"C:\Users\visha\Downloads\IMG_1986 (1).jpg",)
img2 = cv2.imread(r"C:\Users\visha\OneDrive\Desktop\IMG_1986 (1).jpg")
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
methods = ["cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED","cv2.TM_CCORR","cv2.TM_CCORR_NORMED","cv2.TM_CCOEFF","cv2.TM_CCOEFF_NORMED"]
```

```python
width,height,channel = img2.shape
```

```python
for m in methods:
    full_copy = img1.copy()
    #convert string to function use eval 
    method = eval(m)
    res = cv2.matchTemplate(full_copy,img2,method)
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res) # dont change orders 
    if method in[cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left=max_loc
    bottom_right = (top_left[0]+width,top_left[1]+height)
    cv2.rectangle(full_copy,top_left,bottom_right,255,10)
```
