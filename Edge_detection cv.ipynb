```python
blur = cv2.GaussianBlur(image,(5,5),1.4) # x diretion std deviation value . if 0 passed it automatically calculate the sigma y value
plt.imshow(blur,cmap="gray")
```

```python
# 1.median value of image 
#to find best threshold value 
med_val =np.median(image)
med_val
```

```python
lower = int(max(0,0.7* med_val)) #70 % of median
#lower =int(0.7*md_val)
upper = int(min(255,0.3*med_val)) # 30 % of median value 
print(lower,upper)
```

```python
edges = cv2.Canny(blur,threshold1=lower,threshold2=upper+80)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.title("Original")
plt.imshow(image,cmap="gray") # too many edges in this output
plt.subplot(122)
plt.title("edges")
plt.imshow(edges,cmap="gray") 
```
