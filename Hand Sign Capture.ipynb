```python
import cv2
import mediapipe as mp
import numpy as np
import datetime
```

```python
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1)
mp_drawing = mp.solutions.drawing_utils


cap = cv2.VideoCapture(0)
finger_path = []         
drawn_paths = []         
missed_frames = 0         
MAX_MISSED_FRAMES = 15    
hand_visible = True       
tracking_active = False   

def is_only_index_finger_up(hand_landmarks):
    """
    Checks if only the index finger is up and all other fingers are closed.
    """
    index_finger_tip = hand_landmarks.landmark[8]
    middle_finger_tip = hand_landmarks.landmark[12]
    ring_finger_tip = hand_landmarks.landmark[16]
    pinky_tip = hand_landmarks.landmark[20]
    thumb_tip = hand_landmarks.landmark[4]

    index_up = index_finger_tip.y < hand_landmarks.landmark[5].y
    middle_up = middle_finger_tip.y < hand_landmarks.landmark[9].y
    ring_up = ring_finger_tip.y < hand_landmarks.landmark[13].y
    pinky_up = pinky_tip.y < hand_landmarks.landmark[17].y
    thumb_up = thumb_tip.x < hand_landmarks.landmark[3].x  

    return index_up and not (middle_up or ring_up or pinky_up or thumb_up)

def is_five_fingers_up(hand_landmarks):
    """
    Checks if all five fingers are shown.
    """
    index_finger_tip = hand_landmarks.landmark[8]
    middle_finger_tip = hand_landmarks.landmark[12]
    ring_finger_tip = hand_landmarks.landmark[16]
    pinky_tip = hand_landmarks.landmark[20]
    thumb_tip = hand_landmarks.landmark[4]

    return (index_finger_tip.y < hand_landmarks.landmark[5].y and
            middle_finger_tip.y < hand_landmarks.landmark[9].y and
            ring_finger_tip.y < hand_landmarks.landmark[13].y and
            pinky_tip.y < hand_landmarks.landmark[17].y and
            thumb_tip.x < hand_landmarks.landmark[3].x)  


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    
    frame = cv2.flip(frame, 1)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    
    result = hands.process(frame_rgb)

    if result.multi_hand_landmarks:
        hand_visible = True  
        missed_frames = 0  
        for hand_landmarks in result.multi_hand_landmarks:
            
            if is_only_index_finger_up(hand_landmarks):
                
                finger_tip = hand_landmarks.landmark[8]
                h, w, _ = frame.shape
                finger_x, finger_y = int(finger_tip.x * w), int(finger_tip.y * h)

                
                if not tracking_active:
                    tracking_active = True

                
                finger_path.append((finger_x, finger_y))

            elif is_five_fingers_up(hand_landmarks):
                
                save_image = np.zeros_like(frame)  

                
                for path in drawn_paths:
                    for i in range(1, len(path)):
                        cv2.line(save_image, path[i - 1], path[i], (255, 255, 255), 2)  

                for i in range(1, len(finger_path)):
                    cv2.line(save_image, finger_path[i - 1], finger_path[i], (255, 255, 255), 2) 

               
                timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
                cv2.imwrite(f'drawn_paths_{timestamp}.png', save_image)
                print(f"Image saved as drawn_paths_{timestamp}.png")

            else:
                
                tracking_active = False

        
        if tracking_active:
            for i in range(1, len(finger_path)):
                cv2.line(frame, finger_path[i - 1], finger_path[i], (0, 255, 0), 2)  # Draw in green
        else:
            
            drawn_paths.append(finger_path.copy())  
            finger_path = []  

        
        for path in drawn_paths:
            for i in range(1, len(path)):
                cv2.line(frame, path[i - 1], path[i], (0, 255, 0), 2)  # Draw in green

    else:
        hand_visible = False
        missed_frames += 1  
        
        if missed_frames > MAX_MISSED_FRAMES:
            drawn_paths = []
            finger_path = []

    
    cv2.imshow('Air Writing', frame)

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()   
cv2.destroyAllWindows()
```
