```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\visha\OneDrive\Desktop\internal_external.png")
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap="gray")
```

```python
contours,s= cv2.findContours(gray,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
#HEIRARCHY: this is an array that provides information about the image topology 
#contour point: 
```

```python
cv2.drawContours(gray,contours,-1,(255,0,0),2)
# -1 : all the contour point
# color 
#thickness of contour
```

```python
plt.imshow(image,cmap="gray")
```

```python
#shape analysis 
import numpy as np
blank_img = np.zeros((image.shape[0],image.shape[1],3),dtype=np.uint8)
```

```python
cv2.drawContours(blank_img,contours,-1,(160,100,101),5)
```
